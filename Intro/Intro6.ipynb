# MatplotLib

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 10)
print x
```

```python
y_sin = np.sin(x)
y_cos = np.cos(x)
```

```python
%pylab inline
```

```python
plt.plot(x, y_sin, "-r", label = "Sine Wave")
plt.plot(x, y_cos, "-b", label = "Cos Wave")

plt.legend(loc = "upper right")
plt.title("Sine and Cos graphs")
plt.ylim(-1.2, 1.2)
plt.show()
```
